Assignment 5: Tokenization, Stemming, and Lemmatization using NLTK

This assignment demonstrates various tokenization techniques, stemming methods, and lemmatization approaches using the NLTK library.

```python
# Import necessary libraries
import nltk
from nltk.tokenize import WhitespaceTokenizer, WordPunctTokenizer, TreebankWordTokenizer, TweetTokenizer
from nltk.tokenize import MWETokenizer
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import pandas as pd

# Download required NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Sample text for demonstration
sample_text = "Dr. Smith walked into the room and said, 'Hello! How are you doing today?' He wasn't sure if they'd understand his message."
tweet_text = "@john_doe Check out this link: https://example.com #NLP #Python :) It's amazing!!!"

print("Original Text:")
print(sample_text)
print("\nTweet Text:")
print(tweet_text)
```

1. Tokenization Techniques

```python
# 1. Whitespace Tokenization
print("=" * 60)
print("1. WHITESPACE TOKENIZATION")
print("=" * 60)
whitespace_tokenizer = WhitespaceTokenizer()
whitespace_tokens = whitespace_tokenizer.tokenize(sample_text)
print(f"Tokens: {whitespace_tokens}")
print(f"Number of tokens: {len(whitespace_tokens)}")
```

```python
# 2. Punctuation-based Tokenization
print("\n" + "=" * 60)
print("2. PUNCTUATION-BASED TOKENIZATION (WordPunctTokenizer)")
print("=" * 60)
punct_tokenizer = WordPunctTokenizer()
punct_tokens = punct_tokenizer.tokenize(sample_text)
print(f"Tokens: {punct_tokens}")
print(f"Number of tokens: {len(punct_tokens)}")
```

```python
# 3. Treebank Tokenization
print("\n" + "=" * 60)
print("3. TREEBANK TOKENIZATION")
print("=" * 60)
treebank_tokenizer = TreebankWordTokenizer()
treebank_tokens = treebank_tokenizer.tokenize(sample_text)
print(f"Tokens: {treebank_tokens}")
print(f"Number of tokens: {len(treebank_tokens)}")
```

```python
# 4. Tweet Tokenization
print("\n" + "=" * 60)
print("4. TWEET TOKENIZATION")
print("=" * 60)
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(tweet_text)
print(f"Tweet: {tweet_text}")
print(f"Tokens: {tweet_tokens}")
print(f"Number of tokens: {len(tweet_tokens)}")
```

```python
# 5. Multi-Word Expression (MWE) Tokenization
print("\n" + "=" * 60)
print("5. MULTI-WORD EXPRESSION (MWE) TOKENIZATION")
print("=" * 60)

# MWE Tokenizer requires a list of multi-word expressions
mwe = MWETokenizer()
# Add multi-word expressions
mwe.add_mwe(('New', 'York'))
mwe.add_mwe(('machine', 'learning'))

# First tokenize with a basic tokenizer
basic_tokens = sample_text.split()
print(f"Without MWE: {basic_tokens[:10]}...")  # Show first 10

# Then apply MWE tokenizer
# Create a test string with New York
test_text = "I learned machine learning techniques in New York last year."
basic_tokens_test = test_text.split()
mwe_tokens = mwe.tokenize(basic_tokens_test)
print(f"\nTest text: {test_text}")
print(f"Basic tokens: {basic_tokens_test}")
print(f"MWE tokens: {mwe_tokens}")
print(f"Number of MWE tokens: {len(mwe_tokens)}")
```

2. Stemming Techniques

```python
# 1. Porter Stemmer
print("\n" + "=" * 60)
print("1. PORTER STEMMER")
print("=" * 60)

porter_stemmer = PorterStemmer()
words = ['running', 'runs', 'ran', 'walked', 'walking', 'walks', 'computing', 'computers', 'computed']

print(f"{'Original':<15} {'Stem':<15}")
print("-" * 30)
for word in words:
    stem = porter_stemmer.stem(word)
    print(f"{word:<15} {stem:<15}")

# Apply to sample text tokens
print("\nPorter Stemming on sample text (first 10 tokens):")
stemmed_words = [porter_stemmer.stem(word) for word in whitespace_tokens[:10]]
print(stemmed_words)
```

```python
# 2. Snowball Stemmer
print("\n" + "=" * 60)
print("2. SNOWBALL STEMMER")
print("=" * 60)

snowball_stemmer = SnowballStemmer('english')
print(f"{'Original':<15} {'Stem':<15}")
print("-" * 30)
for word in words:
    stem = snowball_stemmer.stem(word)
    print(f"{word:<15} {stem:<15}")

# Apply to sample text tokens
print("\nSnowball Stemming on sample text (first 10 tokens):")
stemmed_words_snowball = [snowball_stemmer.stem(word) for word in whitespace_tokens[:10]]
print(stemmed_words_snowball)

# Compare Porter vs Snowball
print("\nComparison of Porter vs Snowball Stemmer:")
print(f"{'Word':<15} {'Porter':<15} {'Snowball':<15}")
print("-" * 45)
for word in words:
    porter = porter_stemmer.stem(word)
    snowball = snowball_stemmer.stem(word)
    print(f"{word:<15} {porter:<15} {snowball:<15}")
```

3. Lemmatization

```python
# Lemmatization using WordNetLemmatizer
print("\n" + "=" * 60)
print("LEMMATIZATION (WordNet Lemmatizer)")
print("=" * 60)

lemmatizer = WordNetLemmatizer()

# Lemmatization with different part-of-speech tags
extended_words = ['running', 'runs', 'walked', 'walking', 'better', 'best', 'was', 'studies', 'children']

print(f"{'Word':<15} {'Lemma (Default)':<20} {'Lemma (Verb)':<20}")
print("-" * 55)
for word in extended_words:
    lemma_default = lemmatizer.lemmatize(word)
    lemma_verb = lemmatizer.lemmatize(word, pos=wordnet.VERB)
    print(f"{word:<15} {lemma_default:<20} {lemma_verb:<20}")

# Apply to sample text tokens
print("\nLemmatization on sample text (first 10 tokens):")
lemmatized_words = [lemmatizer.lemmatize(word.lower()) for word in whitespace_tokens[:10]]
print(lemmatized_words)
```
