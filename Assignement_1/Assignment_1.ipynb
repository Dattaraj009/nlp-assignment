NLP Assignment 1: Tokenization, Stemming, and Lemmatization

Overview
This assignment demonstrates various tokenization techniques, stemming methods, and lemmatization using the NLTK library.

```python
# Import necessary libraries
import nltk
from nltk.tokenize import (
    word_tokenize, 
    WhitespaceTokenizer,
    RegexpTokenizer,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK data (run once)
print("Downloading NLTK data...")
nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
print("NLTK data downloaded successfully!")

# Sample text for demonstration
sample_text = """Natural Language Processing (NLP) is fascinating! 
It's about understanding human language. We're processing texts, 
tokenizing words, and stemming them. Python is awesome for NLP!!!"""

print("Sample Text:")
print(sample_text)
print("\n" + "="*80 + "\n")
```

1. Whitespace Tokenization
Splits text based on whitespace characters (spaces, tabs, newlines)

```python
whitespace_tokenizer = WhitespaceTokenizer()
whitespace_tokens = whitespace_tokenizer.tokenize(sample_text)

print("WHITESPACE TOKENIZATION")
print(f"Number of tokens: {len(whitespace_tokens)}")
print(f"Tokens: {whitespace_tokens[:15]}...")  # Display first 15 tokens
print()
```

2. Punctuation-Based Tokenization
Uses regular expressions to remove punctuation and tokenize

```python
# Punctuation-based tokenization using RegexpTokenizer
# This tokenizer extracts words while removing punctuation
punct_tokenizer = RegexpTokenizer(r'\w+')
punct_tokens = punct_tokenizer.tokenize(sample_text)

print("PUNCTUATION-BASED TOKENIZATION")
print(f"Number of tokens: {len(punct_tokens)}")
print(f"Tokens: {punct_tokens}")
print()

# Also demonstrate with punctuation included
punct_tokenizer_keep = RegexpTokenizer(r'\w+|[.,!?;]')
punct_tokens_keep = punct_tokenizer_keep.tokenize(sample_text)

print("PUNCTUATION-BASED TOKENIZATION (keeping punctuation)")
print(f"Number of tokens: {len(punct_tokens_keep)}")
print(f"Tokens: {punct_tokens_keep[:20]}...")
print()
```

3. Treebank Tokenization
Follows the Penn Treebank style of tokenization, separating punctuation from words

```python
treebank_tokenizer = TreebankWordTokenizer()
treebank_tokens = treebank_tokenizer.tokenize(sample_text)

print("TREEBANK TOKENIZATION")
print(f"Number of tokens: {len(treebank_tokens)}")
print(f"Tokens: {treebank_tokens[:20]}...")
print()

# Detailed view - show how punctuation is separated
print("Treebank Tokenization (detailed):")
for i, token in enumerate(treebank_tokens[:15], 1):
    print(f"  {i}. '{token}'")
print()
```

4. Tweet Tokenization
Designed for social media text, handles hashtags, mentions, emoticons, and URLs

```python
tweet_tokenizer = TweetTokenizer()

# Sample tweet-like text
tweet_text = "Check out #NLP! @nltk_team is awesome :) Visit https://www.nltk.org for more! #MachineLearning"
tweet_tokens = tweet_tokenizer.tokenize(tweet_text)

print("TWEET TOKENIZATION")
print(f"Tweet text: {tweet_text}")
print(f"Number of tokens: {len(tweet_tokens)}")
print(f"Tokens: {tweet_tokens}")
print()

# Also tokenize the original sample text
sample_tweet_tokens = tweet_tokenizer.tokenize(sample_text)
print("Sample text with Tweet Tokenizer:")
print(f"Number of tokens: {len(sample_tweet_tokens)}")
print(f"Tokens: {sample_tweet_tokens[:20]}...")
print()
```

5. Multi-Word Expression (MWE) Tokenization
Handles multi-word expressions as single tokens

```python
from nltk.tokenize import word_tokenize

# MWE Tokenization
mwe_text = "The United States is a country. New York is a city. Machine Learning is fascinating."

# First, do regular tokenization
regular_tokens = word_tokenize(mwe_text)
print("REGULAR WORD TOKENIZATION (before MWE)")
print(f"Tokens: {regular_tokens}")
print()

# Now apply MWE tokenizer with predefined multi-word expressions
mwe_tokenizer = MWETokenizer([('United', 'States'), ('New', 'York'), ('Machine', 'Learning')])
mwe_tokens = mwe_tokenizer.tokenize(regular_tokens)

print("MWE TOKENIZATION (after applying MWE)")
print(f"Number of tokens: {len(mwe_tokens)}")
print(f"Tokens: {mwe_tokens}")
print()

# Note how "United States", "New York", and "Machine Learning" are now single tokens
print("Notice:")
print("  - 'United' and 'States' are combined into 'United_States'")
print("  - 'New' and 'York' are combined into 'New_York'")
print("  - 'Machine' and 'Learning' are combined into 'Machine_Learning'")
print()
```

6. Porter Stemming
Uses the Porter stemming algorithm to reduce words to their root form

```python
porter_stemmer = PorterStemmer()

# Words to demonstrate stemming
words_to_stem = ['running', 'runs', 'ran', 'runner', 'easily', 'fairly', 'fairly',
                  'processing', 'processed', 'processor', 'tokenization', 'tokenize', 
                  'tokenizing', 'stemming', 'stemmed', 'stems']

print("PORTER STEMMING")
print(f"{'Original Word':<20} {'Stemmed Form':<20}")
print("-" * 40)
for word in words_to_stem:
    stemmed = porter_stemmer.stem(word)
    print(f"{word:<20} {stemmed:<20}")
print()

# Show stemming on sample text tokens
sample_words = word_tokenize(sample_text.lower())
print("Sample text stemmed (first 15 words):")
stemmed_sample = [porter_stemmer.stem(word) for word in sample_words[:15]]
print(f"Original:  {sample_words[:15]}")
print(f"Stemmed:   {stemmed_sample}")
print()
```

7. Snowball Stemming
Snowball is a more sophisticated stemming algorithm supporting multiple languages

```python
snowball_stemmer = SnowballStemmer('english')

print("SNOWBALL STEMMING")
print(f"Available languages: {SnowballStemmer.languages}")
print()

print(f"{'Original Word':<20} {'Snowball Stem':<20} {'Porter Stem':<20}")
print("-" * 60)

for word in words_to_stem:
    snowball_stem = snowball_stemmer.stem(word)
    porter_stem = porter_stemmer.stem(word)
    print(f"{word:<20} {snowball_stem:<20} {porter_stem:<20}")
print()

# Comparison on sample text
print("Comparison: Porter vs Snowball on sample text")
print(f"{'Word':<15} {'Porter':<15} {'Snowball':<15}")
print("-" * 45)
for word in sample_words[5:15]:
    porter = porter_stemmer.stem(word)
    snowball = snowball_stemmer.stem(word)
    print(f"{word:<15} {porter:<15} {snowball:<15}")
print()

# Snowball in other languages
print("Snowball Stemming in Other Languages:")
snowball_french = SnowballStemmer('french')
french_words = ['courir', 'cours', 'couraient']
print(f"French example: {french_words} -> {[snowball_french.stem(w) for w in french_words]}")
print()
```

8. Lemmatization
Uses WordNetLemmatizer to convert words to their base form (lemma) based on POS tags

```python
lemmatizer = WordNetLemmatizer()

print("LEMMATIZATION USING WORDNETLEMMATIZER")
print()

# Lemmatization with different POS tags
lemma_words = ['running', 'runs', 'ran', 'easily', 'better', 'best', 'processing', 'processed']

print("Lemmatization (default - NOUN):")
print(f"{'Word':<15} {'Lemma (NOUN)':<15}")
print("-" * 30)
for word in lemma_words:
    lemma = lemmatizer.lemmatize(word)
    print(f"{word:<15} {lemma:<15}")
print()

# Lemmatization with VERB POS tag
print("Lemmatization (VERB):")
print(f"{'Word':<15} {'Lemma (VERB)':<15}")
print("-" * 30)
for word in lemma_words:
    lemma = lemmatizer.lemmatize(word, pos=wordnet.VERB)
    print(f"{word:<15} {lemma:<15}")
print()

# Lemmatization with ADJ (Adjective) POS tag
print("Lemmatization (ADJ - Adjective):")
adj_words = ['better', 'best', 'easily', 'running', 'processing']
print(f"{'Word':<15} {'Lemma (ADJ)':<15}")
print("-" * 30)
for word in adj_words:
    lemma = lemmatizer.lemmatize(word, pos=wordnet.ADJ)
    print(f"{word:<15} {lemma:<15}")
print()

# Apply lemmatization to sample text
print("Lemmatization on sample text:")
tokens_sample = word_tokenize(sample_text.lower())
lemmatized = [lemmatizer.lemmatize(token, pos=wordnet.VERB) for token in tokens_sample if token.isalpha()][:15]
original = [token for token in tokens_sample if token.isalpha()][:15]

print(f"Original:     {original}")
print(f"Lemmatized:   {lemmatized}")
print()
```
