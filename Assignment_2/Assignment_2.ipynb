NLP Text Representation Techniques

```python
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import seaborn as sns

nltk.download('punkt')
nltk.download('stopwords')

# Sample data - can replace with your own dataset
documents = [
    "natural language processing is fascinating",
    "machine learning and deep learning are powerful",
    "natural language processing uses machine learning",
    "deep learning can process language",
    "text processing with python is great"
]

print("Sample Documents:")
for i, doc in enumerate(documents, 1):
    print(f"{i}. {doc}")
```

1. Bag of Words - Count Occurrence

```python
# Using sklearn's CountVectorizer
count_vect = CountVectorizer(lowercase=True, stop_words='english')
bow_count = count_vect.fit_transform(documents)

# Get feature names (vocabulary)
vocab = count_vect.get_feature_names_out()
print("Vocabulary:", vocab)
print("\nBag of Words - Count Occurrence Matrix:")
print(pd.DataFrame(bow_count.toarray(), columns=vocab))
```

2. Bag of Words - Normalized Count (Relative Frequency)

```python
# Normalize the counts by dividing by total words in each document
bow_array = bow_count.toarray()
bow_normalized = bow_array / bow_array.sum(axis=1, keepdims=True)

print("Normalized Bag of Words (Relative Frequency):")
normalized_df = pd.DataFrame(bow_normalized, columns=vocab)
print(normalized_df.round(4))
```

3. TF-IDF (Term Frequency - Inverse Document Frequency)

```python
# Using sklearn's TfidfVectorizer
tfidf_vect = TfidfVectorizer(lowercase=True, stop_words='english')
tfidf_matrix = tfidf_vect.fit_transform(documents)

print("TF-IDF Matrix:")
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vect.get_feature_names_out())
print(tfidf_df.round(4))

# Show top words by TF-IDF score for each document
print("\n\nTop 3 words per document by TF-IDF score:")
for idx, doc_idx in enumerate(range(len(documents))):
    scores = tfidf_matrix[doc_idx].toarray()[0]
    top_indices = np.argsort(scores)[-3:][::-1]
    print(f"\nDoc {idx + 1}: {documents[doc_idx]}")
    for word_idx in top_indices:
        if scores[word_idx] > 0:
            print(f"  - {tfidf_vect.get_feature_names_out()[word_idx]}: {scores[word_idx]:.4f}")
```

4. Word2Vec Embeddings

```python
# Tokenize documents
tokenized_docs = [word_tokenize(doc.lower()) for doc in documents]

# Train Word2Vec model
# sg=0 uses CBOW, sg=1 uses Skip-gram
w2v_model = Word2Vec(
    sentences=tokenized_docs,
    vector_size=50,  # Dimension of word vectors
    window=3,         # Context window size
    min_count=1,      # Minimum word frequency
    workers=4,        # Number of threads
    sg=1              # 1 for Skip-gram, 0 for CBOW
)

print("Word2Vec Model trained!")
print(f"Vocabulary size: {len(w2v_model.wv)}")
print(f"Vector size: {w2v_model.vector_size}")

# Display word vectors for sample words
print("\n\nWord vectors (first 10 dimensions):")
sample_words = ['natural', 'language', 'learning', 'deep']
for word in sample_words:
    if word in w2v_model.wv:
        print(f"\n{word}: {w2v_model.wv[word][:10]}")
```

```python
# Word similarities
print("Word Similarities using Word2Vec:")
try:
    print(f"\nSimilarity between 'language' and 'processing': {w2v_model.wv.similarity('language', 'processing'):.4f}")
    print(f"Similarity between 'learning' and 'deep': {w2v_model.wv.similarity('learning', 'deep'):.4f}")
except:
    print("Some word pairs not found in vocabulary")

# Most similar words
print("\n\nMost similar words:")
for word in ['language', 'learning']:
    try:
        similar = w2v_model.wv.most_similar(word, topn=2)
        print(f"\nWords similar to '{word}':")
        for similar_word, score in similar:
            print(f"  - {similar_word}: {score:.4f}")
    except:
        print(f"'{word}' not found or not enough similar words")
```
